```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("C:\\Users\\Golden\\Downloads\\homeprices.csv")
df
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color = 'red', marker = '+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3300]])
```

```python
#for different areas in csv file
d = pd.read_csv("C:\\Users\\Golden\\Downloads\\areas.csv")
d
```

```python
reg.predict(d)
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color = 'red', marker = '+')
plt.plot(df.area,reg.predict(df[['area']]), color = 'black')
```

```python
p = reg.predict(d)
d['price'] = p
d
```
